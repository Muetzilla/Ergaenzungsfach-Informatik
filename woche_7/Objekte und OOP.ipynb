### Klassen und Instanzen
- Eine Klasse ist ein Bauplan, der festlegt, welche Eigenschaften ein Objekt hat.

- Eine Instanz einer Klasse ist ein konkretes Objekt, das nach diesem Bauplan erzeugt wird.

- Den Bauplan (die Klasse) gibt es nur einmal.

- Instanzen kann es dagegen viele geben.



#### Beispiel: Eine Datums-Klasse
Wir definieren eine Datumsklasse. Sie soll die Attribute `tag`, `monat` und `jahr` haben.

```python
class Datum:
    def __init__(self, tag, monat, jahr):
        self.tag = tag
        self.monat = monat
        self.jahr = jahr
```

```python
# Zwei Daten erzeugen und verwenden heisst zwei Instanzen zu erzeugen
#
d1 = Datum(1, 9, 2025)  # erste Instanz
d2 = Datum(24, 12, 2025) # zweite Instanz

print(d1.tag, d1.monat, d1.jahr)   # 1 9 2025
print(d2.tag, d2.monat, d2.jahr)   # 24 12 2025
print(d1)
print(d2)
```

```python
# Weitere Daten erzeugen
heute = Datum(1, 9, 2025)
weihnachten = Datum(24, 12, 2025)
silvester = Datum(31, 12, 2025)

# Alle Objekte in eine Liste packen
daten = [heute, weihnachten, silvester]

# Durch die Liste laufen und alle Daten ausgeben
for d in daten:
    print(d.tag, d.monat, d.jahr)

for d in daten:
    print(d)
```

### Dunder-Methoden
Dunder-Methoden sind Funktionen, welche vom System bereit gestellt werden und in speziellen Fällen zum Einsatz kommen.

Der Name kommt von "double-underscore" und ihrer etwas gewöhnungsbedürftigen Schreibweise.

Sie haben bereits die `__init__`-Methode kennengelernt. Im folgenden lernen Sie die `__str__`-Methode kennen.

Dunder-Methoden können überschrieben werden, das heisst, Ihre Funktionsweise kann neu definiert werden.

### Die `__str__` Methode
Die `__str__`-Methode legt fest, wie ein Objekt als menschenlesbarer Text dargestellt wird.

Sie wird automatisch aufgerufen, wenn

- str(obj) aufgerufen wird, dies passiert zum Beispiel, wenn
- print(obj) ausgeführt wird,
- oder das Objekt in einen String eingebettet ist, z. B. `f"Das Datum ist {obj}"`.
- `__str__()`gibt also den "String" zurück, mit der man eine Instanz der Klasse darstellen will.  
- `__str__()` muss explizit implementiert werden. (Wird dies nicht getan - von uns - so ruft Python stattdessen `__repr__()` auf. Auch diese Methode kann überschrieben werden.)

#### Beispiel/Aufgabe: Führe die folgende Zelle und anschliessend die Beispiele oben noch einmal aus. Was hat sich in der Ausgabe mit `print` geändert?

```python
class Datum:
    def __init__(self, tag, monat, jahr):
        self.tag = tag
        self.monat = monat
        self.jahr = jahr

    def __str__(self):
        return f"{self.tag:02d}.{self.monat:02d}.{self.jahr}"
```

```python
for d in [heute, weihnachten, silvester]:
    print(d)
```

Wir wollen uns nur kurz den `__repr__()` ansehen. Er liefert eine  stringbasierte Repräsentation eines Objekts zurück.

```python
for d in [heute, weihnachten, silvester]:
    print(d.__repr__())
```

In diesem Fall sagt er uns, von welcher Klasse das Objekt / die Instanz ist und an welcher Adresse im Speicher man sie findet. Für uns wichtig ist dabei nur: Wir können sehen, dass jede Instanz eine eigene Adresse besitzt. Es handelt sich also wirklich um drei verschiedene Instanzen.

### Das Datums-Modul `datetime`
Unsere Datumsklasse könnten wir erweitern, so dass wir Daten subtrahieren können, oder Tage dazuzählen und so weiter. Dabei wäre es gut, wenn sich das Objekt selbst darum kümmert, ob der Kalender eingehalten wird. Auch mit Schaltjahren.

Für all das gibt es schon ein Modul mit nützlichen Klassen. Schau die die folgenden Beispiele an und ergänze. Hier hat uns jemand viel Arbeit abgenommen. Das wollen Sie nicht selbst programmieren müssen.

```python
from datetime import date, datetime, timedelta
```

##### 1. Heutiges Datum

```python
heute = date.today()
print(heute)              # 2025-09-01
print(heute.day)          # 1
print(heute.month)        # 9
print(heute.year)         # 2025
```

##### 2. Datum selbst erstellen


```python
weihnachten = date(2025, 12, 24)
print(weihnachten)
```

##### 3. Differenz zwischen Daten
Liefert ein neues Objekt, denn die Differenz zwischen zwei Daten kann kein Datum sein. Mit `type` lässt sich der Typ, die Klasse des neuen Objektes bestimmen. Wie heisst sie?

```python
ferien = date(2025, 7, 1)
tage_bis_ferien = ferien - heute
print(tage_bis_ferien.days, "Tage bis zu den Ferien")
print(type(tage_bis_ferien))
```

##### 4. Tage addieren oder subtrahieren

```python
morgen = heute + timedelta(days=1)
gestern = heute - timedelta(days=1)

print("Morgen:", morgen)
print("Gestern:", gestern)
```

##### 5. Schön formattieren (`strftime`)
Die Datumsausgabe kann gesteuert werden.

```python
print(heute.strftime("%d.%m.%Y"))
print(heute.strftime("%A, %d. %B %Y"))
```
