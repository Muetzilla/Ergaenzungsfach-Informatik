```python
text = "WoioW"
textAsChars = list(text)
textReverse = reversed(textAsChars)

if ''.join(textAsChars) == ''.join(textReverse):
  print(f"Der Text {text} ist ein Palidrom!")
else:
  print(f"Der Text {text} ist kein Palidrom!")
```
