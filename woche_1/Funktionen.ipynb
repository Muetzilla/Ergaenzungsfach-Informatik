```python
def relativeFrequency(text, char):
  lengthText = len(text)
  countetChars = text.count(char)
  frequency = countetChars / lengthText * 100
  print(str(frequency) + "%")

relativeFrequency("Lorem ipsum dolor sit amet, consetetur sadipscing elitr, sed diam nonumy eirmod tempor invidunt ut l", "e")
```

```python
def calculateInterest(startCapital, interstRate, interestPeriod, duration):
  capital = startCapital * (1 + interstRate / 100) ** (interestPeriod * duration)
  print(capital)

calculateInterest(1000, 5, 2, 5)
```

```python
def modulo(firstNumber, secondNumber):
  remainer = firstNumber
  counter = 0
  while secondNumber < remainer:
    counter += 1
    remainer = remainer - secondNumber

  return counter


print(f"{modulo(100,3)}")
```
